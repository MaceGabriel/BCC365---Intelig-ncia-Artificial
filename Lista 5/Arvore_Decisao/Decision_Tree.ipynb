```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/rcpsilva/BCC325_ArtificialIntelligence/main/2022-1/Problem%20sets%20(in%20Portuguese)/datasets/decision_tree.csv")
```

```python
df.head()
```

```python
input = df.drop('salary', axis='columns')
target = df['salary']
```

```python
target
```

```python
input
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
input['le_company'] = le_company.fit_transform(input['company'])
input['job_n'] = le_job.fit_transform(input['job'])
input['degree_n'] = le_degree.fit_transform(input['degree'])
```

```python
input
```

```python
inputs_n = input.drop(['company','job','degree'], axis='columns')
```

```python
inputs_n #inputs decodificados
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target) #treinando o modelo com o dataset tratado e os targets
```

```python
model.predict([[2, 2, 1]])
```
